STEP 1: Load and Inspect the Data

```python
import pandas as pd
```

```python
df=pd.read_csv("ICICIBANK.csv")
```

```python
df.head()
```

```python
df.tail()
```

STEP 2: Convert Date Column to Datetime Format

```python
df['Date']=pd.to_datetime(df['Date'])
df=df.sort_values('Date')
df.set_index('Date',inplace=True)
```

STEP 3: Handle Missing or Duplicate Values

```python
print(df.isnull().sum())
df.dropna(inplace=True)
df=df[~df.index.duplicated()]
```

```python
df.drop(['Trades', 'Deliverable Volume', '%Deliverble', 'Symbol', 'Series', 'Prev Close', 'Last', 'Turnover'], axis=1, inplace=True)
```

```python
df.head()
```

STEP 4: Plot the Closing Price to See the Trend

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Closing Price')
plt.title('ICICI Bank Stock Closing Price')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.show()
```

STEP 5: Add Moving Averages (Optional but Insightful)

```python
df['MA20'] = df['Close'].rolling(window=20).mean()
df['MA50'] = df['Close'].rolling(window=50).mean()

df[['Close', 'MA20', 'MA50']].plot(figsize=(14, 7))
plt.title('ICICI Bank - Close Price with 20 & 50 Day Moving Averages')
plt.grid(True)
plt.show()
```

 STEP 6: Save the Cleaned Data

```python
df.to_csv('ICICI_cleaned.csv')
```
