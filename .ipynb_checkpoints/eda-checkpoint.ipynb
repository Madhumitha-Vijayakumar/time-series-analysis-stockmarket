```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('ICICI_cleaned.csv', parse_dates=['Date'], index_col='Date')
```

```python
df.head()
```

1. Line Plot of Closing Prices

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], color='blue', label='Closing Price')
plt.title('ICICI Bank - Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.grid(True)
plt.legend()
plt.show()
```

2. Distribution of Daily Closing Prices

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Close'], bins=50, kde=True, color='green')
plt.title('Distribution of Closing Prices')
plt.xlabel('Closing Price (INR)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

3. Box Plot of Closing Prices (to detect outliers)

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Close'], color='purple')
plt.title('Boxplot of Closing Prices')
plt.xlabel('Closing Price (INR)')
plt.grid(True)
plt.show()
```

4. Monthly Average Closing Prices

```python
monthly_avg = df['Close'].resample('ME').mean()
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg, marker='o', linestyle='-', color='orange')
plt.title('Monthly Average Closing Price')
plt.xlabel('Month')
plt.ylabel('Avg Price (INR)')
plt.grid(True)
plt.show()
```

5. Daily Returns (% Change)

```python
df['Daily Return'] = df['Close'].pct_change() * 100
plt.figure(figsize=(10, 5))
df['Daily Return'].plot(kind='line', color='red')
plt.title('Daily Return Percentage')
plt.ylabel('Return (%)')
plt.grid(True)
plt.show()
```

 6. Correlation Heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
